```python
import random
import tkinter as tk
from tkinter import messagebox
```

```python
def isinword(a, word):
    for b in word:
        if a == b:
            return True
    return False

def create_ans(correct, correct_dict, lives):
    ans = ''
    for letter in correct:        
        if correct_dict[letter] == 1:
            ans += letter + ' '
        else:
            ans += '_ '
    ans += f' Your lives: {lives}'
    return ans

words = ['london', 'moscow', 'belgrade', 'berlin', 'paris', 'brussels', 
         'amsterdam', 'prague', 'ljublana', 'vienna', 'istanbul',
         'stockholm', 'helsinki', 'minsk', 'riga', 'sophia', 'budapest', 
         'athens', 'luxemburg']
```

```python
window = tk.Tk()
window.geometry('500x300') 

frame = tk.Frame(
            master=window,
            relief=tk.RAISED,
            borderwidth=1
        )

frame.pack()

correct = list(random.choice(words))
correct_dict = dict.fromkeys(correct, 0)

lives = 7

mistakes = set()

guess_label = tk.Label(master=frame, text='Enter a letter or a word:')
guess_label.grid(row=0, column=0)

guess_entry = tk.Entry(master=frame)
guess_entry.grid(row=0, column=1)

mistake_label = tk.StringVar()
mistake_label.set("First try")
mistake = tk.Label(window,
                 textvariable = mistake_label)

answer_entry = tk.StringVar()
answer_entry.set(create_ans(correct, correct_dict, lives))
my_label = tk.Label(window,
                 textvariable = answer_entry)
#answer_entry.insert(0, create_ans(correct, correct_dict, lives))

photos = [tk.PhotoImage(file='images/hang0.png'),tk.PhotoImage(file='images/hang2.png'),
         tk.PhotoImage(file='images/hang4.png'),tk.PhotoImage(file='images/hang5.png'),
         tk.PhotoImage(file='images/hang6.png'),tk.PhotoImage(file='images/hang7.png'),
         tk.PhotoImage(file='images/hang9.png'),tk.PhotoImage(file='images/hang11.png')]

imgLabel = tk.Label(window)
imgLabel.config(image=photos[0])
    
def guess():
    global lives
    global answer_entry
    if lives == 0:
        messagebox.showwarning('Hangman', f"You lose! It was {''.join(correct).capitalize()}")
    elif 0 not in correct_dict.values():
        mistake_label.set("Correct! You win!")
    else:
        a = guess_entry.get().lower()
        guess_entry.delete(0, 'end')
    
        if not a.isalpha():
            mistake_label.set("Not a letter")
        else:
            if len(a) > 1:
                if list(a) == correct:
                    for b in correct:
                        correct_dict[b] = 1
                    messagebox.showinfo('Hangman', "Correct! You win!")
                    #answer_entry.delete(0, 'end')
                    answer_entry.set(create_ans(correct, correct_dict, lives))
                    #answer_entry.insert(0, create_ans(correct, correct_dict, lives))
                else:
                    mistake_label.set("Wrong!")
                    lives -= 1
                    #answer_entry.delete(0, 'end')
                    answer_entry.set(create_ans(correct, correct_dict, lives))
                    #answer_entry.insert(0, create_ans(correct, correct_dict, lives))
                    imgLabel.config(image=photos[7 - lives])
            else:
                if isinword(a, correct):
                    if correct_dict[a] == 1:
                        mistake_label.set("Already been")
                    else:
                        mistake_label.set("Correct!")
                        correct_dict[a] = 1

                        #answer_entry.delete(0, 'end')
                        answer_entry.set(create_ans(correct, correct_dict, lives))
                        #answer_entry.insert(0, create_ans(correct, correct_dict, lives))

                        cnt = 1
                        for b in correct:
                            cnt *= correct_dict[b]  
                        if cnt == 1:
                            messagebox.showinfo('Hangman', "Correct! You win!")
                            #answer_entry.delete(0, 'end')
                            answer_entry.set(create_ans(correct, correct_dict, lives))
                            #answer_entry.insert(0, create_ans(correct, correct_dict, lives))
                else:
                    if a in mistakes:
                        mistake_label.set("Already tried")
                    else:
                        mistake_label.set("Wrong!")
                        lives -= 1
                        mistakes.add(a)
                        #answer_entry.delete(0, 'end')
                        answer_entry.set(create_ans(correct, correct_dict, lives))
                        #answer_entry.insert(0, create_ans(correct, correct_dict, lives))
                        imgLabel.config(image=photos[7 - lives])
        if lives == 0:
            messagebox.showwarning('Hangman', f"You lose! It was {''.join(correct).capitalize()}")


def new_game():
    global lives
    global correct
    global correct_dict
    lives = 7
    mistakes = set()
    correct = list(random.choice(words))
    correct_dict = dict.fromkeys(correct, 0)
    answer_entry.set(create_ans(correct, correct_dict, lives))
    imgLabel.config(image=photos[0])
    
my_label.pack()
mistake.pack()
imgLabel.pack()


try_button = tk.Button(master=frame, text='Try', command=guess)
try_button.grid(row=1, column=0)

#window.bind('<Tab>',guess)

new_button = tk.Button(master=frame, text='New game', command=new_game)
new_button.grid(row=1, column=1)

window.mainloop()
        
```
